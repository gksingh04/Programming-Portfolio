```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```python
corona_dataset = pd.read_csv('Covid_19_Confirmed_dataset.csv')
corona_dataset
```

```python
corona_dataset.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
corona_countrywise = corona_dataset.groupby('Country/Region').sum()
```

```python
corona_countrywise.head()
```

#### Task:

#### <li>Calculate the maximum ‘Infection Rate’ for each country and store it in a new column named ‘max_infection_rate’ Create a New Data Frame name ‘Corona Data’ with ‘Country/Region’ as an index and ‘max_infection_rate’ as a column</li>

```python
max_infection_rate = []
for i in corona_countrywise.index:
    max_infection_rate.append(corona_countrywise.loc[i].diff().max())
```

```python
max_infection_rate
```

```python
corona_countrywise['max_infection_rate'] = max_infection_rate
```

```python
corona_countrywise.head()
```

```python
corona_data = pd.DataFrame(corona_countrywise['max_infection_rate'])
```

```python
corona_data.head()
```

#### Task:
##### Visualize the data2 using Folium world map

```python
world = folium.Map(tiles='cartodbpositron',zoom_start=4,max_zoom=7,min_zoom=0,height=500,width='80%')
world
```

```python
len(data2[data2['max_infection_rate']>0].index)
```

```python
for i in range(0,len(data2[data2['max_infection_rate']>0].index)):
    folium.Circle(
        location=[data2.iloc[i]['Lat'],data2.iloc[i]['Long']],
        tooltip = "<h5>"+data2.iloc[i].name+"</h5>"+
        "<li>Max_infection_rate: "+str(data2.iloc[i]['max_infection_rate'])+"</li>"+
        "<li>GDP per capita: "+str(data2.iloc[i]['GDP per capita'])+"</li>"+
        "<li>Social support: "+str(data2.iloc[i]['Social support'])+"</li>"+
        "<li>Healthy life expectancy: "+str(data2.iloc[i]['Healthy life expectancy'])+"</li>"+
        "<li>Freedom to make life choices: "+str(data2.iloc[i]['Freedom to make life choices'])+"</li>",
    
        radius = int(data2.iloc[i]['max_infection_rate'])
).add_to(world)
world
```
